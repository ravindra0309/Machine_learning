```python
# finding absolute value of the correlation
cor_target = abs(corr["MEDV"])

# selecting highly corelated feature 
target_features=cor_target[cor_target>0.3]

#getting the names of the feature
features = [index for index, value in target_features.items()]
features.remove('MEDV')
print(features)
```

```python
from sklearn.preprocessing import StandardScaler
# Select highly correlated features with the target
X = df[features].values
y = df['MEDV'].values

# Scale the input features using StandardScaler
scaler = StandardScaler() # Initialize StandardScaler object
X = scaler.fit_transform(X) # Scale the input features and replace the original X with the scaled version
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=X)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=45)
X_train.shape,y_train.shape 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
R2 = r2_score(y_test, y_pred)
print("mean squared error:",mse)
print("r2_score:",R2)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Y Test (True Values)')
plt.ylabel('Predicted values')
plt.show()

residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
rd=RandomForestRegressor()
rd.fit(X_train,y_train)
y_pred=rd.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
R2 = r2_score(y_test, y_pred)
print("mean squared error:",mse)
print("r2_score:",R2)
```
